```python
import xarray as xr
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
```

```python
path = 'data/climate/cmip6/ssp126/tas_Amon_ACCESS-CM2_ssp126_r8i1p1f1_gn_201501-210012.nc'
```

```python
# Open the dataset
dset = xr.open_dataset(path)
dset
```

```python
fig = plt.figure(figsize=(14,10))
ax = fig.add_subplot(2,1,1,projection=ccrs.PlateCarree())
ax.set_extent([65.0, 107.0, 24.0, 46.0])
dset.tas[10].plot(ax=ax, transform=ccrs.PlateCarree(), cmap='coolwarm')
```

```python
import os
import warnings
warnings.filterwarnings('ignore')
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from cartopy.io.shapereader import Reader
import matplotlib.pyplot as plt
import geopandas as gpd
```

```python
path_hma_gtng = 'data/extents/HMA/gtng_202307_hma_subregions.gpkg'
path_hma_rgi60_lager5 = 'data/glaciers/rgi60_hma_lager5.gpkg'   ### glacier
path_hma_rivers = 'data/rivers/rivers_hma.gpkg'
path_hma_rivers_sword = 'data/rivers/rivers_reaches_hma_sword_v16.gpkg'
path_hma_lakes = 'data/lakes/GLWD_L1_hma.gpkg'
```

```python
### read in data
hma_gtng = gpd.read_file(path_hma_gtng)
hma_rivers = gpd.read_file(path_hma_rivers)
hma_rivers_reaches = gpd.read_file(path_hma_rivers_sword)
hma_glaciers = gpd.read_file(path_hma_rgi60_lager5)
hma_lakes = gpd.read_file(path_hma_lakes)
hma_rivers
```

```python
## rivers (major rivers)
plt.rcParams.update({'font.size':14})
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1,projection=ccrs.PlateCarree())
ax.set_extent([65.0, 107.0, 24.0, 46.0])
ax.gridlines(draw_labels=True, linewidth=0.5, alpha=0.5, zorder=1)
hma_gtng.plot(ax=ax, edgecolor='black', facecolor='none', linewidth=0.5)
hma_rivers.plot(ax=ax, edgecolor='Blue', linewidth=1.5)
# hma_glaciers.plot(ax=ax, edgecolor='yellow', facecolor='yellow', linewidth=1.5)
# hma_lakes.plot(ax=ax, edgecolor='green', linewidth=1.5)
ax.stock_img()
# ax.add_feature(cfeature.LAKES)
# ax.add_feature(cfeature.RIVERS)
```

```python
## rivers (river reaches)
plt.rcParams.update({'font.size':14})
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1,projection=ccrs.PlateCarree())
ax.set_extent([65.0, 107.0, 24.0, 46.0])
ax.gridlines(draw_labels=True, linewidth=0.5, alpha=0.5, zorder=1)
hma_gtng.plot(ax=ax, edgecolor='black', facecolor='none', linewidth=0.5)
hma_rivers_reaches.plot(ax=ax, edgecolor='Blue', linewidth=0.5)
ax.stock_img()
# ax.add_feature(cfeature.LAKES)
# ax.add_feature(cfeature.RIVERS)
```

```python
## glaciers
plt.rcParams.update({'font.size':14})
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1,projection=ccrs.PlateCarree())
ax.set_extent([65.0, 107.0, 24.0, 46.0])
ax.gridlines(draw_labels=True, linewidth=0.5, alpha=0.5, zorder=1)
hma_gtng.plot(ax=ax, edgecolor='black', facecolor='none', linewidth=0.5)
hma_glaciers.plot(ax=ax, edgecolor='yellow', facecolor='yellow', linewidth=1.5)
ax.stock_img()
```

```python
## Lakes
plt.rcParams.update({'font.size':14})
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1,projection=ccrs.PlateCarree())
ax.set_extent([65.0, 107.0, 24.0, 46.0])
ax.gridlines(draw_labels=True, linewidth=0.5, alpha=0.5, zorder=1)
hma_gtng.plot(ax=ax, edgecolor='black', facecolor='none', linewidth=0.5)
hma_lakes.plot(ax=ax, edgecolor='blue', linewidth=1.5)
ax.stock_img()
```

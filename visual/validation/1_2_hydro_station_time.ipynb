```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.dates import YearLocator, DateFormatter
```

```python
# Statistics of all in-situ data
input_path = 'data/ground-observation/water-level-station/hydro_station_lakes_all.csv'
path_save = 'figures/ground-observation/hydro_station_time.png'
df_all = pd.read_csv(input_path)
df_all['date'] = pd.to_datetime(df_all['date'])

# Filter water level data that is not empty.
df_valid = df_all[df_all['water_level'].notna()]
lakes = sorted(df_valid['lake_name'].unique())

plt.figure(figsize=(8, 4))

for i, lake in enumerate(lakes):
    # Filter out all data for the current lake.
    lake_data = df_valid[df_valid['lake_name'] == lake]
    
    dates = lake_data['date'].values
    
    y_positions = np.full(len(dates), i)
    plt.scatter(dates, y_positions, alpha=0.7, s=20, label=lake)
    
    # Print the number of data points for each lake.
    print(f"{lake}: {len(dates)} data points")

plt.yticks(range(len(lakes)), lakes, fontsize=14)  
plt.title('Water Level Over Time', fontsize=16, weight='bold')   
plt.xticks(fontsize=14)  
ax = plt.gca()
# ax.xaxis.set_major_locator(YearLocator())  # One major tick per year.
# ax.xaxis.set_major_formatter(DateFormatter('%Y'))  

plt.grid(True, axis='x', alpha=0.3)
plt.grid(True, axis='y', alpha=0.3)
plt.tight_layout()
# plt.savefig(path_save, dpi=300, bbox_inches='tight', facecolor='white')
```

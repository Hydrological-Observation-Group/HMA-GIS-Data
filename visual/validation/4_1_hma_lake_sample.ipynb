```python
import os
os.chdir('..')
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
path_hma_extent = "data/hma-extent/HMA/hma_gtng_202307_subregions.gpkg"
path_lake_valid = "visual/validation/valid_lake_s2.gpkg" 

hma = gpd.read_file(path_hma_extent)
lake_valid = gpd.read_file(path_lake_valid)
```

```python
# 可视化展示
fig, ax = plt.subplots(figsize=(10, 8))

hma.boundary.plot(
    ax=ax, 
    edgecolor="black", 
    linewidth=0.6, 
    label="HMA boundary"
)

lake_valid.plot(
    ax=ax, 
    color="blue",       
    marker="o",       
    markersize=10,     
    label="Lake validation points",
    alpha=0.8           
)

ax.legend()
ax.set_xlabel("Longitude")
ax.set_ylabel("Latitude")
ax.grid(True, linestyle="--", alpha=0.4)
ax.set_title("Lake Validation Points in HMA")

plt.show()
```

```python
# 4. 保存和展示
fig.savefig("figures/validation/lake_validation_loc.png", bbox_inches='tight', dpi=200)

```

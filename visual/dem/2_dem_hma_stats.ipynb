### dem statistics.


```python
import numpy as np
import geopandas as gpd
import cartopy.crs as ccrs 
import matplotlib.pyplot as plt
from utils.scatter_plot import scatter_plot
```

```python
path_hma_gtng = 'data/hma-extent/HMA/hma_gtng_202307_subregions.gpkg'
path_dem_stats = 'data/dem/dem_stats.gpkg'
```

```python
## read hma and glacier extent
dem_stats_gdf = gpd.read_file(path_dem_stats)
hma_gtng_gdf = gpd.read_file(path_hma_gtng)
dem_stats_gdf.head(3)
```

```python


name_subregions = [name.split(' (')[0] for name in dem_stats_gdf['full_name'].values] 

fig, ax = plt.subplots(figsize=(10, 4))
ax.bar(
    x=name_subregions,
    height=dem_stats_gdf['mean'],
    yerr=dem_stats_gdf['std'],
    width=0.5,
    color="#2D0878",
    capsize=4,
    linewidth=0.5,
    ecolor = "#A70C0C",  # color for error bars
    alpha=0.7,
    )
ax.set_xticks(np.arange(len(name_subregions)), labels=name_subregions, rotation=30, ha='right')
ax.set_xlabel('Region')
ax.set_ylabel('Mean elevation (m)')
ax.set_title('Mean DEM elevation with standard deviation by region')
# fig.savefig('figures/dem/hma_subregions_dem_mean_std.png', bbox_inches='tight', dpi=300)
```

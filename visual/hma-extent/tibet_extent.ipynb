### **Visualize the gis data on the high-mountain asia region.**

```python
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib.pyplot as plt
import geopandas as gpd
import matplotlib.patches as mpatches
```

## 1. HMA extent
#### 1.1 HMA extent from existing studies; 


```python
path_hma_brun2017 = 'data/extents/HMA/hma_brun2017_subregions.gpkg'
path_hma_holch2019 = 'data/extents/HMA/hma_bolch2019_subregions.gpkg'
path_hma_liu2022 = 'data/extents/HMA/hma(Pan_TPlateau)_liu2022_subregions.gpkg'
path_hma_gtng = 'data/extents/HMA/gtng_202307_hma_subregions.gpkg'
```

```python
hma_brun2017 = gpd.read_file(path_hma_brun2017)
hma_holch2019 = gpd.read_file(path_hma_holch2019)
hma_liu2022 = gpd.read_file(path_hma_liu2022)
hma_gtng = gpd.read_file(path_hma_gtng)
hma_holch2019.head()
```

```python
plt.rcParams.update({'font.size':12})
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1, projection=ccrs.PlateCarree())
ax.set_extent([65.0, 107.0, 24.0, 46.0])
ax.gridlines(draw_labels=True, linewidth=0.5, alpha=0.5, zorder=1)
hma_brun2017.plot(ax=ax, edgecolor='green', facecolor='none', linewidth=1.5)
hma_holch2019.plot(ax=ax, edgecolor='brown', facecolor='none', linewidth=1.5)
hma_liu2022.plot(ax=ax, edgecolor='blue', facecolor='none', linewidth=1.5)
hma_gtng.plot(ax=ax, edgecolor='yellow', facecolor='none', linewidth=1.5)
ax.stock_img()
ax.add_feature(cfeature.LAKES)
ax.add_feature(cfeature.RIVERS)
legend_patches = [
  mpatches.Patch(edgecolor='green', facecolor='none', label='brun2017'),
  mpatches.Patch(edgecolor='brown', facecolor='none', label='holch2019'),
  mpatches.Patch(edgecolor='blue', facecolor='none', label='liu2022'),
  mpatches.Patch(edgecolor='yellow', facecolor='none', label='gtng')]
ax.legend(handles=legend_patches, loc='lower left', labelspacing=0.8)
ax.set_title('hma subregions', fontsize=14)
# plt.savefig('figures/hma_subregions_compare.png', bbox_inches='tight', dpi=100)

```

#### 1.2. Tibet Plateau region from Chinese TPDC.

```python
### TP boundary from Chinese TPDC
path_tb_2500 = 'data/extents/TPlateau/TPBoundary_2500.gpkg'
path_tb_3000 = 'data/extents/TPlateau/TPBoundary_3000.gpkg'
path_tb_hf = 'data/extents/TPlateau/TPBoundary_HF.gpkg'
path_tb_new2021 = 'data/extents/TPlateau/TPBoundary_new2021.gpkg'
```

```python
tb_2500 = gpd.read_file(path_tb_2500)
tb_3000 = gpd.read_file(path_tb_3000)
tb_hf = gpd.read_file(path_tb_hf)
tb_new2021 = gpd.read_file(path_tb_new2021)
tb_new2021.head()
```

```python
plt.rcParams.update({'font.size':12})
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1,projection=ccrs.PlateCarree())
# ax.set_extent([62.0, 107.0, 24.0, 43.0])
ax.set_extent([62.0, 107.0, 23.0, 44.0])
ax.gridlines(draw_labels=True, linewidth=0.5, alpha=0.5, zorder=1)
tb_2500.plot(ax=ax, edgecolor='MediumBlue', facecolor='none', linewidth=1.5)
tb_3000.plot(ax=ax, edgecolor='green', facecolor='none', linewidth=1.5)
tb_hf.plot(ax=ax, edgecolor='Purple', facecolor='none', linewidth=1.5)
tb_new2021.plot(ax=ax, edgecolor='Red', facecolor='none', linewidth=1.5)
ax.stock_img(); ax.coastlines()
legend_patches = [
  mpatches.Patch(edgecolor='MediumBlue', facecolor='none', label='TPBoundary_2500'),
  mpatches.Patch(edgecolor='green', facecolor='none', label='TPBoundary_3000'),
  mpatches.Patch(edgecolor='Purple', facecolor='none', label='TPBoundary_HF'),
  mpatches.Patch(edgecolor='Red', facecolor='none', label='TPBoundary_new2021')]
ax.legend(handles=legend_patches, loc='lower left', labelspacing=0.7, fontsize=10)

ax.set_title('TP boundaries', fontsize=14)
plt.savefig('figures/TP_boundaries.png', bbox_inches='tight', dpi=100)
```

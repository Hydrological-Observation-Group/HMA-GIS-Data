### hma and tibet extent visualization

```python
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib.pyplot as plt
import geopandas as gpd
import matplotlib.patches as mpatches
```

#### 1. HMA extent

```python
path_hma_brun2017 = 'data/hma-extent/HMA/hma_brun2017_subregions.gpkg'
path_hma_holch2019 = 'data/hma-extent/HMA/hma_bolch2019_subregions.gpkg'
path_hma_liu2022 = 'data/hma-extent/HMA/hma(Pan_TPlateau)_liu2022_subregions.gpkg'
path_hma_gtng = 'data/hma-extent/HMA/hma_gtng_202307_subregions.gpkg'
```

```python
hma_brun2017 = gpd.read_file(path_hma_brun2017)
hma_holch2019 = gpd.read_file(path_hma_holch2019)
hma_liu2022 = gpd.read_file(path_hma_liu2022)
hma_gtng = gpd.read_file(path_hma_gtng)
hma_holch2019.head()
```

```python
plt.rcParams.update({'font.size':12})
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw={'projection': ccrs.PlateCarree()})
ax.set_extent([55, 117.0, 16, 48.0])

### plot hma boundary
hma_brun2017.plot(ax=ax, edgecolor='green', facecolor='none', linewidth=1)
hma_holch2019.plot(ax=ax, edgecolor='brown', facecolor='none', linewidth=1)
hma_liu2022.plot(ax=ax, edgecolor='blue', facecolor='none', linewidth=1)
hma_gtng.plot(ax=ax, edgecolor='yellow', facecolor='none', linewidth=1)

## plot auxiliary data
ax.add_feature(cfeature.LAND, facecolor='WhiteSmoke')
ax.add_feature(cfeature.OCEAN)
ax.add_feature(cfeature.COASTLINE)


## add legend
legend_patches = [
  mpatches.Patch(edgecolor='green', facecolor='none', label='brun2017'),
  mpatches.Patch(edgecolor='brown', facecolor='none', label='bolch2019'),
  mpatches.Patch(edgecolor='blue', facecolor='none', label='liu2022'),
  mpatches.Patch(edgecolor='yellow', facecolor='none', label='gtng')]
ax.legend(handles=legend_patches, loc=(0.78, 0.75), labelspacing=0.7, fontsize=10)

ax.gridlines(draw_labels=True, linewidth=0.5, alpha=0.5)
ax.set_title('hma subregions', fontsize=14)
# plt.savefig('figures/hma-extent/hma_extent.png', bbox_inches='tight', dpi=200)
```

#### 2. Tibet Plateau (TP) extent
the data is from Chinese TPDC.

```python
### TP boundary from Chinese TPDC
path_tb_2500 = 'data/hma-extent/TPlateau/TPBoundary_2500.gpkg'
path_tb_3000 = 'data/hma-extent/TPlateau/TPBoundary_3000.gpkg'
path_tb_hf = 'data/hma-extent/TPlateau/TPBoundary_HF.gpkg'
path_tb_new2021 = 'data/hma-extent/TPlateau/TPBoundary_new2021.gpkg'
```

```python
tb_2500 = gpd.read_file(path_tb_2500)
tb_3000 = gpd.read_file(path_tb_3000)
tb_hf = gpd.read_file(path_tb_hf)
tb_new2021 = gpd.read_file(path_tb_new2021)
tb_new2021.head()
```

```python
plt.rcParams.update({'font.size':12})
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw={'projection': ccrs.PlateCarree()})
ax.set_extent([60.0, 110.0, 18.0, 47.0])

### plot tp boundary
tb_2500.plot(ax=ax, edgecolor='MediumBlue', facecolor='none', linewidth=1.5)
tb_3000.plot(ax=ax, edgecolor='green', facecolor='none', linewidth=1.5)
tb_hf.plot(ax=ax, edgecolor='Purple', facecolor='none', linewidth=1.5)
tb_new2021.plot(ax=ax, edgecolor='Red', facecolor='none', linewidth=1.5)

## plot auxiliary data
ax.add_feature(cfeature.LAND, facecolor='WhiteSmoke')
ax.add_feature(cfeature.OCEAN)
ax.add_feature(cfeature.COASTLINE)

## add legend
legend_patches = [
  mpatches.Patch(edgecolor='MediumBlue', facecolor='none', label='TPBoundary_2500'),
  mpatches.Patch(edgecolor='green', facecolor='none', label='TPBoundary_3000'),
  mpatches.Patch(edgecolor='Purple', facecolor='none', label='TPBoundary_HF'),
  mpatches.Patch(edgecolor='Red', facecolor='none', label='TPBoundary_new2021')]
ax.legend(handles=legend_patches, loc=(0.65, 0.77), labelspacing=0.7, fontsize=10)

ax.gridlines(draw_labels=True, linewidth=0.5, alpha=0.5, zorder=1)
ax.set_title('TP boundaries', fontsize=14)
# plt.savefig('figures/hma-extent/TP_extent.png', bbox_inches='tight', dpi=200)
```

```python
import os
import warnings
warnings.filterwarnings('ignore')
import cartopy.crs as ccrs # type: ignore
import cartopy.feature as cfeature
from cartopy.io.shapereader import Reader
import matplotlib.pyplot as plt
import geopandas as gpd
```

```python
path_hma_gtng = 'data/extents/HMA/gtng_202307_hma_subregions.gpkg'
path_hma_rgi60_lager2 = 'data/glaciers/rgi60_hma_larger2.gpkg'   ### glacier
```

```python
### read in data
hma_gtng = gpd.read_file(path_hma_gtng)
hma_glaciers = gpd.read_file(path_hma_rgi60_lager2)
```

```python
## glaciers
plt.rcParams.update({'font.size':14})
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1,projection=ccrs.PlateCarree())
ax.set_extent([65.0, 107.0, 24.0, 46.0])
ax.gridlines(draw_labels=True, linewidth=0.5, alpha=0.5, zorder=1)
hma_gtng.plot(ax=ax, edgecolor='black', facecolor='none', linewidth=0.5)
hma_glaciers.plot(ax=ax, edgecolor='yellow', facecolor='yellow', linewidth=1.5)
ax.stock_img()
ax.set_title('Distribution of RGI6.0 glaicers (area > 2km2)', fontsize=14)
plt.savefig('figures/hma_glaciers.png', bbox_inches='tight', dpi=100)
```

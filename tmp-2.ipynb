```python
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
```

```python
path = 'data/climate/cmip6/pr_Amon_CanESM5_ssp126_r22i1p1f1_gn_201501-210012.nc'
path_2 = 'data/climate/cmip6/pr_Amon_NESM3_ssp126_r2i1p1f1_gn_201501-210012.nc'
```

```python
# Open the dataset
dset = xr.open_dataset(path)
dset
```

```python
# Open the dataset
dset_2 = xr.open_dataset(path_2)
dset_2
```

```python
fig = plt.figure(figsize=(14,10))
ax = fig.add_subplot(2,1,1,projection=ccrs.PlateCarree())
# ax.set_extent([65.0, 107.0, 24.0, 46.0])
dset.pr[10].plot(ax=ax, transform=ccrs.PlateCarree(), cmap='coolwarm', vmin=0, vmax=0.0002)

ax2 = fig.add_subplot(2,1,2,projection=ccrs.PlateCarree())
# ax.set_extent([65.0, 107.0, 24.0, 46.0])
dset_2.pr[10].plot(ax=ax2, transform=ccrs.PlateCarree(), cmap='coolwarm', vmin=0, vmax=0.0002)

```

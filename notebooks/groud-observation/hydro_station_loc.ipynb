### Create in-situ station location file.

```python
import os
import geopandas as gpd
from shapely.geometry import Point 
```

```python
lake_name = ['Zhari Namco',  'DawaCo', 'DazegCo', 'BamCo', 'PengCo',
              'NamCo', 'SilingCo', 'MemarCo', 'Luotuo lake', 'LumajiangdongCo',
              'Jieze Caka', 'Kalakuli']
lake_coords = [(85.61, 30.93), (84.96, 31.24), (87.53, 31.89), (90.58, 31.26), (90.97, 31.50),
               (90.60, 30.74), (88.63, 30.95), (82.31, 34.22), (81.86, 34.46), (81.61, 34.02),
               (80.90, 33.95), (75.03, 38.43)]
lake_gdf = gpd.GeoDataFrame({'lake_name': lake_name},
                            geometry=[Point(xy) for xy in lake_coords],
                            crs="EPSG:4326")
```

```python
path_save = 'data/ground-observation/hydro_station.gpkg'
if os.path.exists(path_save): os.remove(path_save) 
lake_gdf.to_file(path_save, driver='GPKG')
```
